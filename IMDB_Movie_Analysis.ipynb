```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/imdb_top_1000.csv')
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Gross'] = df['Gross'].replace({',': ''}, regex=True)
df['Gross'] = df['Gross'].replace(np.nan, 0)
df['Gross'] = df['Gross'].astype(int)
```

```python
df.describe()
```

```python
top_votes = df.sort_values(['No_of_Votes'], ascending = False)
```

```python
fig,ax=plt.subplots(figsize=(40,10))
sns.barplot(x=top_votes['Series_Title'][:10], y=top_votes['No_of_Votes'][:10], palette = 'hls')
plt.title('Top 10 Voted Movies', fontweight = 'bold', fontsize = 20)
plt.xlabel('Movies', fontsize = 10, fontweight = 'bold')
plt.ylabel('Votes', fontsize = 10, fontweight = 'bold')
plt.show()
```

```python
top_gross = df.sort_values(['Gross'], ascending = False)
```

```python
fig, ax = plt.subplots(figsize = (40,10))
sns.barplot(x=top_gross['Series_Title'][:10], y=top_gross['Gross'][:10], palette = 'hls')
plt.title('Top 10 Gross Movies', fontsize = 15, fontweight = 'bold')
plt.xlabel('Movies', fontsize = 10, fontweight = 'bold')
plt.ylabel ('Gross', fontsize = 10, fontweight = 'bold')
plt.show()
```

```python
plt.figure(figsize=(30,20))

plt.xticks(rotation = 90)

plt.bar(x = df[df['IMDB_Rating']>=8.0]['Genre'],height= df[df['IMDB_Rating']>=8.0]['Gross'])
plt.title('Movie Genre-Gross Revenue', fontsize = 15, fontweight = 'bold')
plt.xlabel('Genre')
plt.ylabel('Gross Revenue')
plt.show()
```
